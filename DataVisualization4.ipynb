```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
tips = sns.load_dataset('tips')
tips.head()
```

```python
plt.scatter(tips['total_bill'],tips['tip'], c = tips['sex'].map({"Male" : "r", "Female" : "b"}), s = tips['size']* 30)
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.title('Total Bill vs Tip')
plt.show()
```

```python
scatterplot = sns.scatterplot(x = "total_bill", y = "tip", hue = "sex", style = "smoker", data = tips, size = "size")
scatterplot.set_title('Total Bill vs Tip')
plt.show()
```

```python
sns.relplot(kind = "line", x = "total_bill", y = "tip", hue = "sex", style = "smoker", size = "size", data = tips, col = "sex", row = "smoker")
plt.show()
```

```python
sns.displot(data=tips,x='total_bill',kde=True,col="sex",kind="hist")
sns.rugplot(data=tips,x='total_bill',color="g")
plt.show()
```

**Iris Dataset**

```python
sns.set_style('white')
iris = sns.load_dataset('iris')
iris.head()
```

```python
sns.scatterplot(data= iris, x="sepal_length",y="sepal_width",hue="species")
```

```python
sns.scatterplot(data=iris,x="petal_length",y="petal_width",hue="species")
plt.show()
```

```python
sns.relplot(data=iris,x="petal_length",y="petal_width",hue="species",kind="line")
plt.show()
```

```python
sns.displot(data=tips,x="total_bill",y="tip",kind="hist", hue = "sex", col = "day", col_wrap = 2)
plt.show()
```

```python
g = sns.FacetGrid(data = tips, col="day", hue = "smoker", col_wrap = 2)
g.map(sns.histplot, "total_bill", "tip")
g.add_legend()
plt.show()
```

```python
g = sns.FacetGrid(data = tips, col="day", hue = "smoker", col_wrap = 2)
g.map(sns.kdeplot, "total_bill", "tip")
g.add_legend()
plt.show()
```

```python
g = sns.FacetGrid(data = tips, col="day", hue = "smoker", col_wrap = 2)
g.map(sns.scatterplot, "total_bill", "tip")
g.add_legend()
plt.show()
```

```python
g = sns.FacetGrid(data = tips, col="day", hue = "smoker", col_wrap = 2)
g.map(sns.displot, "total_bill", "tip")
g.add_legend()
plt.show()
```

```python
g = sns.FacetGrid(data = tips, col="day", hue = "smoker", col_wrap = 2)
g.map(sns.rugplot, "total_bill", "tip")
g.add_legend()
plt.show()
```
