```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
tips = sns.load_dataset('tips')
tips.head()
```

```python
plt.scatter(tips['total_bill'],tips['tip'], c = tips['sex'].map({"Male" : "r", "Female" : "b"}), s = tips['size']* 30)
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.title('Total Bill vs Tip')
plt.show()
```

```python
scatterplot = sns.scatterplot(x = "total_bill", y = "tip", hue = "sex", style = "smoker", data = tips, size = "size")
scatterplot.set_title('Total Bill vs Tip')
plt.show()
```

```python
sns.relplot(kind = "line", x = "total_bill", y = "tip", hue = "sex", style = "smoker", size = "size", data = tips, col = "sex", row = "smoker")
plt.show()
```

```python
sns.displot(data=tips,x='total_bill',kde=True,col="sex",kind="hist")
sns.rugplot(data=tips,x='total_bill',color="g")
plt.show()
```
