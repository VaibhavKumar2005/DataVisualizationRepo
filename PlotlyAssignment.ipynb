```python
import plotly.express as px
import pandas as pd
import numpy as np

# Create 5-minute time intervals for full day
time_intervals = pd.date_range("2025-01-01 00:00", "2025-01-01 23:55", freq="5min")

# Example CPU usage values (can be replaced with real data)
np.random.seed(42)
cpu_auth = np.random.randint(10, 90, len(time_intervals))
cpu_pay = np.random.randint(15, 95, len(time_intervals))
cpu_rec = np.random.randint(5, 85, len(time_intervals))

# Prepare DataFrame
df = pd.DataFrame({
    "Time": time_intervals,
    "Auth": cpu_auth,
    "Payments": cpu_pay,
    "Recommendations": cpu_rec
})

# Convert to long format for Plotly
df_melt = df.melt(
    id_vars="Time",
    value_vars=["Auth", "Payments", "Recommendations"],
    var_name="Microservice",
    value_name="CPU Usage (%)"
)

# Plotly multi-line chart
fig = px.line(
    df_melt,
    x="Time",
    y="CPU Usage (%)",
    color="Microservice",
    markers=True,
    title="CPU Usage of Microservices at 5-minute Intervals"
)

fig.update_layout(
    xaxis_title="Time (5-minute Intervals)",
    yaxis_title="CPU Usage (%)",
    hovermode="x unified",
    template="plotly_dark"
)

fig.show()
```
