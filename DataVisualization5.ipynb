```python
data = {
    'University':['IIT Delhi','IIT Bombay','IISc Bangalore','IIT Madras','IIT Kanpur'],
    'latitude': [28.545955,19.1334,13.0218597,12.9916,26.5123],
    'longitude': [77.18614,72.9133,77.5671423, 80.2337,80.2329],
    'Students' : [11000, 10500, 16000, 5000, 7000]
}
df = pd.DataFrame(data)
df
```

```python
m = folium.Map(location=[20, 77], zoom_start=4)

for name, lat, lon, students in zip(df['University'], df['latitude'], df['longitude'], df['Students']):
    color = 'green' if students > 10000 else 'orange' if students >= 8000 else 'red'
    folium.Marker(
        [lat, lon],
        popup=f"<b>{name}</b><br>Students:{students}",
        icon=folium.Icon(color=color)
    ).add_to(m)

m
```
