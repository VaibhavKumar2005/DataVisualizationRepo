```python
import plotly.express as px
```

```python
df = px.data.stocks()
df
```

```python
import plotly.express as px
px.line(data_frame = df,x="date", y =["GOOG", "AAPL", "MSFT","NFLX","FB","AMZN"], title = 'Stock Price Over Time')
```

```python
fig = px.line(df,x="date", y =["GOOG", "AAPL", "MSFT","NFLX","FB","AMZN"], title = 'Stock Prices Over Time')

fig.show()
```

```python
fig.update_layout(
    xaxis = dict(
        dtick = "M1",
        tickformat = "%b %Y",
        #tickformat = '%Y-%m-%d',
        tickangle=45
    )
)
```

```python
fig.update_layout(
    xaxis = dict(
        dtick = "M4",
        tickformat = "%b %Y",
        #tickformat = '%Y-%m-%d',
        tickangle=45
    )
)
```

```python
fig = px.line(df,x="date", y =["GOOG", "AAPL", "MSFT","NFLX","FB","AMZN"], title = 'Stock Prices Over Time')
fig.update_traces(mode='markers+lines')
fig.show()
```

```python
fig = px.line(df,x="date", y =["GOOG", "AAPL", "MSFT","NFLX","FB","AMZN"],
              title = 'Stock Prices Over Time',
              labels={'date':'Date','value':'Price(USD)'},
              template='plotly_dark')
fig.update_traces(mode='markers+lines')
fig.update_layout(width=900, height=500, legend_title_text = 'Company', title_x=0.5)

fig.show()
```

```python
fig.update_traces(hovertemplate='Date: %{x}<br>Price: %{y:$,.2f}')

fig.show()
```

# **Scatter Plot**

```python
iris=px.data.iris()

fig = px.scatter(
    iris,
    x = "sepal_width",
    y = "sepal_length",
    title = "Iris: Sepal Length vs Width",
)

fig.show()
```

```python
fig.write_html("iris_plot.html")
```

```python
fig = px.scatter(
    iris, x="sepal_width", y="sepal_length",
    color="species",
    size = "petal_length",
    symbol = "species",
    hover_name="species",
    hover_data=["petal_width"],
    title="Add Size, Symbol, Hover",
)

fig.show()
```

```python
fig_3d = px.scatter_3d(
    iris, x="sepal_width", y="sepal_length", z="petal_length",
    color="species",
    title="Iris 3D Structure",
)

fig.show()
```

```python
tips = px.data.tips()

fig = px.bar(
    tips,
    x = "day",
    y = "total_bill",
    title = "Total Bill by Day",
)

fig.show()
```

```python
df_avg = tips.groupby("day")["total_bill"].mean()

fig = px.bar(
    tips,
    x = "day",
    y = "total_bill",
    title = "Total Bill by Day",
    text="total_bill"
)

fig.show()
```

```python
df_stack = tips.groupby(["day", "smoker"], as_index=False)["total_bill"].sum()

fig = px.bar(
    df_stack,
    x = "day",
    y = "total_bill",
    color = "smoker",
    barmode = "group",
    title = "Total Bill by Day (Stacked By Smoker)"
)

fig.show()
```
