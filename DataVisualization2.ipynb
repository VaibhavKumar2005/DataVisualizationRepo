```python
import matplotlib.pyplot as plt
employees= ["Ram","Shytam","Mohan","Radha"]
matric1= [80,7,6,8]
matric2= [60,80,70,85]
yesrs= [2022,2023,2024,2025]
```

```python
plt.figure(figsize=(4,6))
plt.plot(yesrs,matric2,marker="o")
plt.title("matric2 over the years")
plt.xlabel("Years")
plt.ylabel("Matric 2 (0 to 100)")
plt.xticks(yesrs,rotation=90)
```

```python
plt.barh(yesrs,matric2,height=0.6,color=["r","g","b","pink"])
plt.yticks(yesrs)
plt.title("matric2 over the years")
plt.xlabel("Matric 2 (0 to 100)")
plt.ylabel("Years")
plt.show()
```

```python
plt.bar(yesrs,matric2,width=0.6,color=["r","g","b","pink"])
plt.xticks(yesrs)
plt.title("matric2 over the years")
plt.ylabel("Matric 2 (0 to 100)")
plt.xlabel("Years")
plt.show()
```

```python
fig,(hbar,vbar)=plt.subplots(1,2)
hbar.barh(yesrs,matric2,height=0.6,color=["r","g","b","pink"])
hbar.set_title("matric2 over the years")
hbar.set_xlabel("matric2")
hbar.set_ylabel("Years")

vbar.bar(yesrs,matric2,width=0.6,color=["r","g","b","pink"])
hbar.set_title("matric2 over the years")
hbar.set_ylabel("matric2")
hbar.set_xlabel("Years")
```

```python
fig, (hbar, vbar) = plt.subplots(1, 2, figsize=(12, 5))

hbar.barh(yesrs, matric2, color=["r", "g", "b", "pink"])
hbar.set_title("Horizontal Bar - Metric 2 over Years")
hbar.set_xlabel("Metric 2")
hbar.set_ylabel("Years")

vbar.bar(yesrs, matric2, width=0.6, color=["r", "g", "b", "pink"])
vbar.set_title("Vertical Bar - Metric 2 over Years")
vbar.set_xlabel("Years")
vbar.set_ylabel("Metric 2 (0 to 100)")

plt.show()
```

```python
plt.figure(figsize=(10,8))
plt.subplot(1,2,1)
plt.bar(yesrs,matric2,width=0.6,color=["r","g","b","pink"])

plt.title("matric2 over the years")
plt.ylabel("Matric 2 (0 to 100)")
plt.xlabel("Years")

plt.subplot(1,2,2)
plt.bar(yesrs,matric2,width=0.6,color=["r","g","b","pink"])
for i, v in enumerate(matric2):
    #print(i,v)
    plt.text(yesrs[i], v + 2, str(v), ha="center", fontweight="bold")
plt.title("matric2 over the years")
plt.ylabel("Matric 2 (0 to 100)")
plt.xlabel("Years")
plt.xticks(yesrs)
plt.show()
```

```python
employees= ["Ram","Shytam","Mohan","Radha","Ananda"]
matric1= [8,7,6,8,7]
matric2= [60,80,70,85,70]
matric3= [600,700,555,850,556]
yesrs= [2022,2023,2024,2025,2026]
```

```python
import numpy as np

x = np.arange(len(employees))
print(x)
width = 0.25
```

```python
x1 = x - width
x2 = x
x3 = x + width
```

```python
plt.bar(x=x2,height=matric1,width=0.25,color=["g","r","pink","b"])
plt.bar(x=x1,height=matric2,width=0.25,color=["g","r","pink","b"])
plt.bar(x=x2,height=matric3,width=0.25,color=["g","r","pink","b"])
plt.xticks(x,employees)
plt.title("Employees")
plt.ylabel("Matric 2 (0 to 100)")
plt.xlabel("Employees")
plt.show()
```

```python
plt.bar(x=yesrs,height=matric1,width=0.2,color=["g","r","pink","b"])

plt.bar(x=yesrs,height=matric1,width=0.2,color=["g","r","pink","b"])

plt.title("matric2 over the years")
plt.ylabel("Matric 2 (0 to 100)")
plt.xlabel("Years")
```

```python
import seaborn as sns
tips = sns.load_dataset('tips')
daywisetotalbill = tips.groupby('day')[["total_bill", "tip","size"]].sum().reset_index()
x = np.arange(len(daywisetotalbill["day"]))
x2 = x
x1 = x - .25
x3 = x + .25
plt.bar(x1, daywisetotalbill["total_bill"], width= 0.25, color = "g")
plt.bar(x2, daywisetotalbill["tip"], width= 0.25, color = "r")
plt.bar(x3, daywisetotalbill["size"], width= 0.25, color = "b")
plt.xticks(x, daywisetotalbill["day"])
plt.title("Day-wise Total Bill, Tip, and Size")
plt.xlabel("Day")
plt.ylabel("Amount")
plt.show()

#plt.bar(tips['day'], tips["total_bill"])
# sns.get_dataset_names()
```
